```python
train = pd.read_csv("/kaggle/input/feedback-prize-2021/train.csv")
train.head(10)
```

```python

train['discourse_type'].value_counts().sort_values(ascending=True).plot(kind='barh', width=0.8, figsize=(8,8))
plt.tick_params(right=False, labelright=False)
plt.box(False)
ylocs, ylabs = plt.xticks()
for i, v in enumerate(train['discourse_type'].value_counts().sort_values(ascending=True)):
    if i == 0:
        plt.text(v - 4000, i, str(v), color='white', weight='bold', fontsize=11)
    elif i == 1:
        plt.text(v - 4600, i, str(v), color='white', weight='bold', fontsize=11)
    else:
        plt.text(v - 6000, i, str(v), color='white', weight='bold', fontsize=11)
plt.yticks(fontsize=16, rotation=0)
plt.tight_layout()

plt.show()
```

```python
document_id = train['id'].unique()[1]
document_df = train[train['id'] == document_id]


with open(os.path.join('/kaggle/input/feedback-prize-2021/train', document_id + '.txt')) as f:
    content = f.read()
f.close()

highlighted_text = []

color_dict = {
    'Lead': 'rgb(252, 151, 144)',
    'Position': 'rgb(253, 255, 143)', 
    'Evidence': 'rgb(164, 250, 132)', 
    'Claim': 'rgb(126, 237, 233)', 
    'Concluding Statement': 'rgb(150, 206, 255)',
    'Counterclaim': 'rgb(255, 120, 163)', 
    'Rebuttal':'rgb(247, 109, 111)'
}

for i, word in enumerate(content.split()):
    
    for j, prediction_string in enumerate(document_df.iloc[:, 7]):
        prediction_string = [int(k) for k in prediction_string.split(' ')]
        if i in prediction_string:
            highlighted_text.append('<span style="background-color:' + color_dict[document_df.iloc[j,5]] + ';">' + word + '</span>')  
        
        if j == len(document_df.iloc[:, 7]):
            highlighted_text.append(word)
            
            
        
from IPython.display import HTML
        
highlighted_text = ' '.join(highlighted_text)

display(HTML(highlighted_text))

legend_html = []
for category in train.iloc[:, 5].unique():
    legend_html.append('<span style="background-color:' + color_dict[category] + ';"><b>' + category + '</b></span>')
    
print('\n\n')

display(HTML(' '.join(legend_html)))
```
